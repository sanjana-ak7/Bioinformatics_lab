```python
!pip install biopython
from Bio import Entrez
from Bio.Blast import NCBIWWW, NCBIXML

Entrez.email = "your.email@example.com"  # Provide your email here

sequence_ids = ["NC_000017.11", "NT_033777.3"]

for seq_id in sequence_ids:
    print(f"Fetching sequence for ID: {seq_id}")
    # Fetch sequence in FASTA format from NCBI
    handle = Entrez.efetch(db="nucleotide", id=seq_id, rettype="fasta", retmode="text")
    fasta_data = handle.read()
    handle.close()

    # Extract only the sequence string (remove FASTA header and newlines)
    lines = fasta_data.split('\n')
    seq_str = ''.join(line.strip() for line in lines if not line.startswith('>'))

    print(f"Performing BLAST for sequence ID: {seq_id}")
    # Perform the BLAST search with sequence string
    result_handle = NCBIWWW.qblast(
        "blastn",  # BLAST program (nucleotide blast)
        "nt",      # Database (nucleotide database)
        seq_str    # Query sequence string
    )

    # Parse the BLAST results
    blast_records = NCBIXML.parse(result_handle)

    # Print some information about the hits
    for blast_record in blast_records:
        print(f"Query: {blast_record.query}")
        for alignment in blast_record.alignments:
            for hsp in alignment.hsps:
                print(f"  Alignment: {alignment.title}")
                print(f"  Length: {alignment.length}")
                print(f"  E-value: {hsp.expect}")
                print(f"  Query start: {hsp.query_start}, Query end: {hsp.query_end}")
                print(f"  Subject start: {hsp.sbjct_start}, Subject end: {hsp.sbjct_end}")
                print(f"  Score: {hsp.score}")
                print(f"  Identities: {hsp.identities}")
                print(f"  Gaps: {hsp.gaps}")
                print(f"  Query: {hsp.query[0:75]}...")
                print(f"  Match: {hsp.match[0:75]}...")
                print(f"  Sbjct: {hsp.sbjct[0:75]}...")
                print("-" * 80)

    result_handle.close()
```

```python
# =============================
#  Global, Local, and Multiple Sequence Alignment (p53 gene)
# Using Biopython + Clustal Omega
# =============================

# Step 1: Install dependencies
!pip install biopython --quiet
!apt-get install -y -qq clustalo

from Bio import Entrez, SeqIO, AlignIO, pairwise2
from Bio.Align import AlignInfo
from itertools import combinations
import os, textwrap

# Step 2: User input
Entrez.email = "your.email@example.com"  # replace with your email
accession_ids = ["NM_000546.6", "NM_011640.3", "XM_016922764.2"]  # Human, Mouse, Chimp (p53 gene)

outdir = "/content/p53_alignment_results"
os.makedirs(outdir, exist_ok=True)

# Step 3: Fetch sequences from NCBI
seq_records = []
for acc in accession_ids:
    try:
        print(f"Fetching {acc} ...")
        handle = Entrez.efetch(db="nucleotide", id=acc, rettype="fasta", retmode="text")
        rec = SeqIO.read(handle, "fasta")
        handle.close()
        seq_records.append(rec)
        print(f" Got {rec.id} | length = {len(rec.seq)}")
    except Exception as e:
        print(f" Could not fetch {acc}: {e}")

if not seq_records:
    raise SystemExit("No sequences fetched. Check IDs or internet connection.")

# Save sequences to FASTA
fasta_path = os.path.join(outdir, "p53_sequences.fasta")
SeqIO.write(seq_records, fasta_path, "fasta")
print("\nSequences saved to:", fasta_path)

# Step 4: Pairwise Global and Local alignments
def pct_identity(a, b):
    matches, length = 0, 0
    for x, y in zip(a, b):
        if x == "-" and y == "-": continue
        length += 1
        if x == y: matches += 1
    return (matches / length * 100) if length else 0

# Scoring parameters
match, mismatch, gap_open, gap_extend = 2, -1, -2, -0.5

print("\n===  GLOBAL and LOCAL Alignments ===")
for s1, s2 in combinations(seq_records, 2):
    print(f"\n{s1.id}  vs  {s2.id}")

    # Global (Needleman-Wunsch)
    g = pairwise2.align.globalms(str(s1.seq), str(s2.seq), match, mismatch, gap_open, gap_extend, one_alignment_only=True)[0]
    g_pid = pct_identity(g.seqA, g.seqB)
    print(f" Global Alignment: Score = {g.score:.1f}, Identity = {g_pid:.2f}%")
    print(textwrap.fill(g.seqA[:100], 80))
    print(textwrap.fill(g.seqB[:100], 80))

    # Local (Smith-Waterman)
    l = pairwise2.align.localms(str(s1.seq), str(s2.seq), match, mismatch, gap_open, gap_extend, one_alignment_only=True)[0]
    l_pid = pct_identity(l.seqA, l.seqB)
    print(f" Local Alignment:  Score = {l.score:.1f}, Identity = {l_pid:.2f}%")
    print(textwrap.fill(l.seqA[:100], 80))
    print(textwrap.fill(l.seqB[:100], 80))

# Step 5: Multiple Sequence Alignment (Clustal Omega)
msa_out = os.path.join(outdir, "p53_msa.fasta")
!clustalo -i $fasta_path -o $msa_out --force --outfmt=fasta --threads=2

alignment = AlignIO.read(msa_out, "fasta")
summary = AlignInfo.SummaryInfo(alignment)
consensus = summary.dumb_consensus(threshold=0.7, ambiguous="N")

print(f"\n===  Multiple Sequence Alignment ===")
print(f"Sequences aligned: {len(alignment)}")
print(f"Alignment length: {alignment.get_alignment_length()}")
print("Consensus sequence (70% agreement):\n", consensus)

# Print a small preview of MSA
for rec in alignment:
    print(f"\n>{rec.id}")
    print(rec.seq[:200])

print("\n All alignments complete. Files saved in:", outdir)
```

```python
!pip install Bio
from Bio import Entrez
from Bio.Blast import NCBIWWW
from Bio.Blast import NCBIXML
# Set your email address for Entrez
Entrez.email = "your.email@example.com" # Replace with your email

sequence_ids = ["NC_000017.11", "NT_033777.3"]

# Define a range for the sequences (example: first 1000 bases)
# You'll need to adjust the start and end points based on the actual region you're interested in
query_range_start = 1
query_range_end = 1000

# Perform BLAST for each sequence
for seq_id in sequence_ids:
    print(f"Performing BLAST for sequence ID: {seq_id}")
    print(f"  Using range: {query_range_start} to {query_range_end}")
    # Perform the BLAST search
    result_handle = NCBIWWW.qblast(
        "blastn",  # BLAST program (nucleotide blast)
        "nt",      # Database (nucleotide database)
        seq_id,    # Query sequence ID
        query_from=query_range_start,
        query_to=query_range_end
    )

    # Parse the BLAST results
    blast_records = NCBIXML.parse(result_handle)

    # Print some information about the hits
    for blast_record in blast_records:
        print(f"Query: {blast_record.query}")
        if blast_record.alignments:
            for alignment in blast_record.alignments:
                for hsp in alignment.hsps:
                    print(f"  Alignment: {alignment.title}")
                    print(f"  Length: {alignment.length}")
                    print(f"  E-value: {hsp.expect}")
                    print(f"  Query start: {hsp.query_start}, Query end: {hsp.query_end}")
                    print(f"  Subject start: {hsp.sbjct_start}, Subject end: {hsp.sbjct_end}")
                    print(f"  Score: {hsp.score}")
                    print(f"  Identities: {hsp.identities}")
                    print(f"  Gaps: {hsp.gaps}")
                    print(f"  Query: {hsp.query[0:75]}...")
                    print(f"  Match: {hsp.match[0:75]}...")
                    print(f"  Sbjct: {hsp.sbjct[0:75]}...")
                    print("-" * 80)
        else:
            print("  No alignments found for this query.")


    result_handle.close()
```

```python
def read_fasta(file_path):
    """Reads a FASTA file and returns a dictionary of sequences."""
    sequences = {}
    current_sequence_name = None
    with open(file_path, 'r') as f:
        for line in f:
            line = line.strip()
            if line.startswith('>'):
                current_sequence_name = line[1:]
                sequences[current_sequence_name] = ''
            elif current_sequence_name:
                sequences[current_sequence_name] += line
    return sequences

def simple_sequence_comparison(seq1, seq2, match_score=1, mismatch_penalty=-1, gap_penalty=-2):
    """Performs a simple character-by-character comparison of two sequences."""
    score = 0
    alignment_seq1 = ""
    alignment_seq2 = ""
    i, j = 0, 0

    while i < len(seq1) and j < len(seq2):
        if seq1[i] == seq2[j]:
            score += match_score
            alignment_seq1 += seq1[i]
            alignment_seq2 += seq2[j]
            i += 1
            j += 1
        else:
            # Simple approach: penalize mismatches and advance both
            score += mismatch_penalty
            alignment_seq1 += seq1[i]
            alignment_seq2 += seq2[j]
            i += 1
            j += 1

    # Add remaining characters with gap penalty
    while i < len(seq1):
        score += gap_penalty
        alignment_seq1 += seq1[i]
        alignment_seq2 += '-'
        i += 1
    while j < len(seq2):
        score += gap_penalty
        alignment_seq1 += '-'
        alignment_seq2 += seq2[j]
        j += 1

    return score, alignment_seq1, alignment_seq2

# Example usage:
# Create dummy FASTA files for demonstration
with open('sequence1.fasta', 'w') as f:
    f.write('>SeqA\n')
    f.write('ATGCGTACGTAC')

with open('sequence2.fasta', 'w') as f:
    f.write('>SeqB\n')
    f.write('ATGCGTAGCTAC')

# Read sequences from FASTA files
sequences1 = read_fasta('sequence1.fasta')
sequences2 = read_fasta('sequence2.fasta')

# Assuming each file has only one sequence for this example
seq1_name = list(sequences1.keys())[0]
seq2_name = list(sequences2.keys())[0]
sequence1 = sequences1[seq1_name]
sequence2 = sequences2[seq2_name]

# Perform simple comparison
score, aligned_seq1, aligned_seq2 = simple_sequence_comparison(sequence1, sequence2)

print(f"Sequence 1 ({seq1_name}): {sequence1}")
print(f"Sequence 2 ({seq2_name}): {sequence2}")
print("\nSimple Comparison Results:")
print(f"Alignment Score: {score}")
print(f"Aligned Sequence 1: {aligned_seq1}")
print(f"Aligned Sequence 2: {aligned_seq2}")
```
