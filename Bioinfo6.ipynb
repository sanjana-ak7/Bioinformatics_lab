```python
# Sample DNA sequence (e.g., a plasmid or a gene)
dna_sequence = "GAATTCGCATGCTAGCTAGCATGTAGCATGCTAGCTAGCATGTAGCATGCCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATCGATTCGACGGATCCGTCGATGAATTC"

# Define a dictionary of restriction enzymes and their recognition sites
# Key: Enzyme Name, Value: Recognition Sequence (5' to 3')
restriction_enzymes = {
    "EcoRI": "GAATTC",
    "HindIII": "AAGCTT",
    "BamHI": "GGATCC",
    "PstI": "CTGCAG",
    "SmaI": "CCCGGG"
}

# Function to get the reverse complement of a DNA sequence
def reverse_complement(seq):
    complement = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    return "".join([complement[base] for base in seq[::-1]])

# Function to find all occurrences of a pattern in a sequence
def find_all_occurrences(sequence, pattern):
    occurrences = []
    start = 0
    while True:
        start = sequence.find(pattern, start)
        if start == -1:
            break
        occurrences.append(start)
        start += 1  # Move past this occurrence to find the next
    return occurrences

print(f"Analyzing DNA sequence of length: {len(dna_sequence)}\n")

# Iterate through each enzyme and find its sites
for enzyme_name, recognition_site in restriction_enzymes.items():
    print(f"--- {enzyme_name} ({recognition_site}) ---")

    # Search on the forward strand
    forward_sites = find_all_occurrences(dna_sequence, recognition_site)
    if forward_sites:
        print(f"  Forward strand sites found at positions: {forward_sites}")
    else:
        print("  No sites found on the forward strand.")

    # Search on the reverse complement strand
    rev_comp_site = reverse_complement(recognition_site)
    reverse_sites = find_all_occurrences(dna_sequence, rev_comp_site)
    if reverse_sites:
        # Adjust positions to be relative to the original DNA sequence
        # A site at index 'x' on the reverse complement corresponds to len(dna_sequence) - x - len(rev_comp_site) on the forward strand
        # For simplicity, we'll report the start index of the reverse complement match
        print(f"  Reverse complement sites ({rev_comp_site}) found at positions: {reverse_sites}")
    else:
        print("  No sites found on the reverse complement strand.")
    print("\n")

```
